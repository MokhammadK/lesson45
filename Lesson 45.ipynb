```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

data = sm.datasets.sunspots.load_pandas().data

start_year = 1700
end_year = 2008
years = np.arange(start_year, end_year + 1)
dates = pd.date_range(start=f"{start_year}-01-01", end=f"{end_year}-12-31", freq='A')
ts = pd.Series(data['SUNACTIVITY'].values, index=dates)

plt.figure(figsize=(12, 6))
plt.plot(ts)
plt.title('Yearly Sunspots Data 1700-2008')
plt.xlabel('Year')
plt.ylabel('Sunspots Count')
plt.show()

model = sm.tsa.ARIMA(ts, order=(5, 1, 0))  
results = model.fit()

forecast_years = 22
forecast = results.forecast(steps=forecast_years)

plt.figure(figsize=(12, 6))
plt.plot(ts, label='Observed Data')
plt.plot(pd.date_range(start='2009', periods=forecast_years, freq='A'), forecast, label='Forecast', color='red')
plt.title('Yearly Sunspots Data Forecast 2009-2030')
plt.xlabel('Year')
plt.ylabel('Sunspots Count')
plt.legend()
plt.show()
```

```python
forecast_years = 22
forecast = results.forecast(steps=forecast_years)

print("     2009  2030 :")
for year, value in zip(range(2009, 2031), forecast):
    print(f" {year}: {value:.2f}")
```
